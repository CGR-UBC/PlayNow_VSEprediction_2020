# Supplementary Table 1: Descriptive Statistics

# libraries

```python
import pandas as pd
import numpy as np
```

# parameter

```python
input_path = ''
ouput_path = ''
```

# data

```python
data = pd.read_csv(input_path)
columns = list(data)[1:-1]
```

# descriptive statistics

```python
overall_mean = []
overall_std = []
vse_mean = []
vse_std = []
control_mean = []
control_std = []

for col in columns: 
    overall_mean.extend([data[col].mean()])
    overall_std.extend([data[col].std()])
    vse_mean.extend([data.loc[data.exclusion==1,col].mean()])
    vse_std.extend([data.loc[data.exclusion==1,col].std()])
    control_mean.extend([data.loc[data.exclusion==0,col].mean()])
    control_std.extend([data.loc[data.exclusion==0,col].std()])  
    
df_desc_stats = pd.DataFrame(data = list(zip(overall_mean,
                                             overall_std,
                                             vse_mean,
                                             vse_std,
                                             control_mean,
                                             control_std)),
                             index = columns,
                             columns = ['overall_mean', 
                                        'overall_std',
                                        'vse_mean', 
                                        'vse_std',
                                        'control_mean', 
                                        'control_std'])
```
